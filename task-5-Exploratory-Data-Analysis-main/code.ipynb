```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
# --------------------
# A. Handle missing values
# --------------------

# 1. Fill 'Embarked' with the most common value (mode)
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# 2. Fill 'Age' with median value
df['Age'] = df['Age'].fillna(df['Age'].median())

# --------------------
# B. Convert data types
# --------------------
df['Pclass'] = df['Pclass'].astype('category')
df['Survived'] = df['Survived'].astype('category')

# --------------------
# C. Feature Engineering
# --------------------

# 1. Extract Title from Name
df['Title'] = df['Name'].str.extract(r',\s*([^\.]+)\.')

# 2. Create FamilySize
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1

# 3. Create IsAlone flag
df['IsAlone'] = (df['FamilySize'] == 1).astype(int)

# 4. Create FareBand (quartiles)
df['FareBand'] = pd.qcut(df['Fare'], 4, labels=False)

# 5. Create AgeBand (custom bins)
df['AgeBand'] = pd.cut(df['Age'], bins=[0, 12, 20, 40, 60, 120], labels=False)

# --------------------
# D. Verify no missing values remain
# --------------------
print(df.isnull().sum())

# --------------------
# E. Save cleaned dataset
# --------------------
import os
os.makedirs("../data/processed", exist_ok=True)
df.to_csv("../data/processed/train_clean.csv", index=False)
```

```python
numeric_cols = ['Age', 'Fare', 'SibSp', 'Parch']

for col in numeric_cols:
    plt.figure(figsize=(6,3))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[col] < Q1 - 1.5*IQR) | (df[col] > Q3 + 1.5*IQR)]
    print(f"{col}  {len(outliers)} outliers")
```

```python
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
IQR = Q3 - Q1
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

df['Fare'] = np.where(df['Fare'] > upper_limit, upper_limit,
                      np.where(df['Fare'] < lower_limit, lower_limit, df['Fare']))
```

```python
sns.histplot(df['Fare'], kde=True)
plt.title("Fare before log transform")
plt.show()

df['Fare_log'] = np.log1p(df['Fare'])  # log1p handles 0 values

sns.histplot(df['Fare_log'], kde=True)
plt.title("Fare after log transform")
plt.show()
```
